```
#computing face encodings for train images:
import face_recognition

Aishwarya_encoding=face_recognition.face_encodings(Aishwarya)[0]
Na_Mo_encoding=face_recognition.face_encodings(Na_Mo)[0]
obama_encoding=face_recognition.face_encodings(obama)[0]
trump_encoding=face_recognition.face_encodings(trump)[0]
dhoni_encoding=face_recognition.face_encodings(ms_dhoni)[0]
```

```
encodings=[Aishwarya_encoding,Na_Mo_encoding,obama_encoding,trump_encoding,dhoni_encoding]
```

```
encodings=np.array(encodings)
encodings.shape
```

```
#Loading test images
test1=cv2.imread("Aishwarya_Rai2.jpg")
test1=cv2.cvtColor(test1,cv2.COLOR_BGR2RGB)

test2=cv2.imread("Barack_Obama2.jpg")
test2=cv2.cvtColor(test2,cv2.COLOR_BGR2RGB)

test3=cv2.imread("Donald_Trump2.jpg")
test3=cv2.cvtColor(test3,cv2.COLOR_BGR2RGB)

test4=cv2.imread("MS Dhoni2.jpg")
test4=cv2.cvtColor(test4,cv2.COLOR_BGR2RGB)

test5=cv2.imread("NaMo2.jpg")
test5=cv2.cvtColor(test5,cv2.COLOR_BGR2RGB)

```

```
#finding the encodings of the test images and making a list of all encodings further converting into numpy array
enc1=face_recognition.face_encodings(test1)[0]
enc2=face_recognition.face_encodings(test2)[0]
enc3=face_recognition.face_encodings(test3)[0]
enc4=face_recognition.face_encodings(test4)[0]
enc5=face_recognition.face_encodings(test5)[0]

test_encodings=[enc1,enc2,enc3,enc4,enc5]
```

```
test_encodings=np.array(test_encodings)
```

```
print(type(test_encodings))
print(test_encodings.shape)
```

```
from scipy.spatial import distance as d
```

```
#Our test images have only single face hence our code will be :   
```

```
for test_encod in test_encodings:
  results=[]
  for encod in encodings:
    dist=d.euclidean(encod,test_encod)
    results.append(dist)
  threshold=0.6    #if the threshold is less than 0.6, it means our image has been mis classified hence will be stored as false in results array
  results=np.array(results) <=threshold

  if(results[0]):
    plt.imshow(test1)
    plt.show()
    print("Aishwarya")
  elif(results[1]):
    plt.imshow(test5)
    plt.show()
    print("Narendra Modi") 
  elif(results[2]):
    plt.imshow(test2)
    plt.show()
    print("Obama")
  elif(results[3]):
    plt.imshow(test3)
    plt.show()
    print("Donald Trump")
  elif(results[4]):
    plt.imshow(test4)
    plt.show()
    print("MS Dhoni")      


```
